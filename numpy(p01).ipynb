```python
# COPIAS vs VISTAS

# Demostraci√≥n de que las variables pueden ser referencias al mismo objeto en memoria.
aa = np.array([2, 3, 7, 8, 5])
bb = aa  # bb es una referencia al mismo arreglo que aa
print(id(bb) == id(aa))  #True # bb es una referencia al mismo arreglo que aa
bb[2] = 33  # Modificar el tercer elemento a trav√©s de bb
print(aa)  # [ 2  3 33  8  5]  # aa refleja el cambio hecho a trav√©s de bb  
aa[0] = 0  # Modificar el primer elemento a trav√©s de aa
print(bb)  # [ 0  3 33  8  5]  # bb refleja el cambio hecho a trav√©s de aa
print(bb is aa)  #True # ambas variables apuntan al mismo objeto en memoria.

print("------------")
# Creando una copia independiente del arreglo
cc = bb.copy()  # cc es una copia independiente de aa
cc[2] = 55  # Modificar el tercer elemento de cc
print(cc is bb)  # False  # cc es una copia independiente de bb
print(bb)  # [ 0  3 33  8  5]  # bb no se ve afectado por el cambio en cc
print(cc)  # [ 0  3 55  8  5]  # cc es independiente de bb
print("------------")


# Por defecto, las operaciones de slicing y las funciones que crean nuevos arreglos devuelven vistas (views) del arreglo original, no copias independientes.
# Una vista es una referencia al mismo bloque de memoria que el arreglo original, por lo que los cambios en la vista afectan al original y viceversa.
# Si deseas crear una copia independiente del arreglo, puedes usar el m√©todo .copy().



# b = a[1:4]  # Vista del arreglo original (elementos desde el √≠ndice 1 hasta el 3)
# b[0] = 99   # Modificar el primer elemento de la vista
# print(a)  # [ 2 99  7  8  5]
# # Para crear una copia independiente:
# d = a[1:4].copy()  # Copia independiente del arreglo original
# d[0] = 55          # Modificar el primer elemento de la copia
# print(a)  # [ 2 99  7  8  5]                
```

```python
# Creacion y comportamiento de vistas en NumPy

aa = np.array([2, 3, 7, 8, 5])
bb = aa  ## bb es un alias o una referencia al objeto aa. Son exactamente el mismo objeto en memoria
# Formas de crear una vista:
cc = aa.view()  # bb es una vista del arreglo aa
cc = aa[:]  # bb es una vista del arreglo aa
print("------")
print(id(bb) == id(aa)) # True # bb es una referencia al mismo arreglo que aa
print(id(cc) == id(aa)) # False # cc es una vista, no una referencia directa (pero tampoco es una copia independiente)
print("------")
vw1_a = aa.view()  # vw_a es una vista del arreglo aa
vw2_a = aa.view()  # vw_a es una vista(2) del arreglo aa
vw1_b = bb.view()  # vw_a es una vista del arreglo bb (que es una referencia a aa)
# nota pero aa y bb siguen siendo referencias al mismo arreglo
print(vw1_a is aa)  # False  # vw_a es una vista, no una referencia directa
print(vw1_b is bb)  # False  # vw_a es una vista, no una referencia directa
# ...
print("--- Modificando el arreglo original ---")
aa[-1] = 99  # Modificar el √∫ltimo elemento del arreglo original
# Afecta a todas las vistas y referencias al mismo bloque de memoria
print(f"aa: {aa}")  
print(f"bb: {bb}") 
print(f"vw1_a: {vw1_a}")
print(f"vw2_a: {vw2_a}")
print(f"vw1_b: {vw1_b}")

print("--- Modificando una vista ---")
vw1_b[-2] = 55  
# Afecta a todas las vistas y referencias al mismo bloque de memoria
print(f"aa: {aa}")  
print(f"bb: {bb}") 
print(f"vw1_a: {vw1_a}")
print(f"vw2_a: {vw2_a}")
print(f"vw1_b: {vw1_b}")

# La principal ventaja de las vistas es la eficiencia, especialmente en operaciones complejas de reestructuraci√≥n de datos, sin necesidad de copiar gigabytes de informaci√≥n.
# Las vistas ahorran una cantidad masiva de memoria en comparaci√≥n con una copia completa (.copy()).

# Copia Completa (.copy()): Crea un bloque de datos nuevo en otra parte de la RAM. (Ocupa el doble de espacio).
# Vista (.view()): Solo crea un peque√±o encabezado nuevo con metadatos. El espacio extra es insignificante (unos pocos bytes).
```

```python
# Reasignaci√≥n de variables y referencias en NumPy
# Cuidado: reasignar la variable a un nuevo objeto rompe la referencia original.
aa = np.array([2, 3, 7, 8, 5])
bb = aa  # bb es una referencia al mismo arreglo que aa
print(id(bb) == id(aa)) # True
print("------")
aa = np.array([2, 3, 7, 8, 5])
print(id(bb) == id(aa)) # falso
# aa deja de apuntar al objeto original y apunta a un objeto completamente nuevo.
```

```python
# Pero, manipular un elemento es posible:
print("Modificar un elemento del arreglo original:")
print(a)  # [2 3 7 8 5]
a[2] = 99 # Modificar el tercer elemento del arreglo original
print(a)  # [ 2  3 99  8  5]
print("------------")
# Indexaci√≥n B√°sica (Slicing)
# Como tambien acceder a elementos (## üü¶ Indexaci√≥n y slicing)
print(a[0])    # primer elemento
print(a[-1])   # √∫ltimo elemento
print(a[1:4])  # elementos desde el √≠ndice 1 hasta el 3 (excluye el 4) 
print(a[:3])   # primeros 3 elementos (incluye 0, 1, 2)
print(a[2:])   # desde el √≠ndice 2 hasta el final
print(a[::-1])    # arreglo invertido
```

```python
###########CUIDADO CON LAS VISTAS (views) EN NUMPY ###########
# Referencias en el caso de slices (slicing) en NumPy
# Cuando haces un slice (selecci√≥n de rango) en un array de NumPy, por defecto, tambi√©n se devuelve una vista:
ar = np.array([7, 1, 9, 6, 3])
ars = ar[:3]
print("Original ar  :", ar)    # [7 1 9 6 3]
print("Slice ars  :", ars)   # [7 1 9]
print("Demostraion de que ars es una vista de ar (ars.base):", ars.base)
    # Demostraion de que ars es una vista de ar (ars.base): [7 1 9 6 3]
print("---")
print("El valor de id(ar) es igual al valor de id(ars.base):", id(ar) == id(ars.base)) # True  # Mismas direcciones de memoria
print("ar y ars.base apuntan al mismo objeto:", ar is ars.base) # True # is sirve para comparar si dos variables apuntan al mismo objeto en memoria
# NO HACER ESTO: # print(id(ar) is id(ars.base))  # Resultado: False ### RESULTADO NO ESPERADO
# RAZ√ìN:
# 1. 'id(objeto)' se llama dos veces, creando dos objetos de tipo 'int' (n√∫meros).
# 2. El operador 'is' comprueba si son el MISMO objeto en memoria, no si tienen el mismo valor.
# 3. Aunque los VALORES de los ID son id√©nticos (porque ar y ars.base son el mismo array),
#    Python trata las dos llamadas a 'id()' como dos OBJETOS enteros DISTINTOS.

print("---")
# Pero ar y ars comparten la misma memoria para esos tres primeros elementos.
ars[1] = 22
print("Modified ar origin :", ar)
print("Modified ars[1]=22 :", ars)
print("---")
ar[0] = 11
print("Modified ar[0]=11 :", ar)
print("Modified ars view :", ars)
print("---")

### NOTA: En las listas normales de Python no ocurre lo mismo: el slice de una lista es una copia superficial, no una vista.
print("IMPORTANTE: lista normal en Python no comparte memoria en slices:")
l = [7, 1, 9, 6, 3]
ls = l[:3]      # slice de lista
ls[1] = 22
print(l)   # [7, 1, 9, 6, 3]  ‚Üê no cambia
print(ls)  # [7, 22, 9]
```

```python
# Subarreglos mediante Fancy Indexing (Selecci√≥n no Contigua) #  [[]]
# Acceso a elementos (indexing) 
z = np.array([2, 3, 77, 8, 55, 10, 6, 1, 44, 9])
sub_z = z[[2, 4, 5, 8]] # Acceder a elementos en las posiciones 2 y 4
print(sub_z)  # [77 55 10 44]   # Subarreglo con los elementos en las posiciones 2 y 4
sub2_z = z[[True, True, False, True, False, False, True, True, False, True]]
print(sub2_z)  # [ 2  3  8  6  1  9]   # Subarreglo con los elementos donde el √≠ndice booleano es True
print("---- M√°scara booleana ----")
# sub3_z = z[[True, False*8, True]] # ERROR
# Crea la lista (como mascara): [True] + [False, False, ..., False] (8 veces) + [True]
mascara_correcta = [True] + [False] * 8 + [True]
sub3_z = z[mascara_correcta]
print(sub3_z)  # [2 9]   # Subarreglo con los elementos donde el √≠ndice booleano es True
# fancy indexing es una t√©cnica para acceder a m√∫ltiples elementos de un arreglo utilizando listas o arreglos de √≠ndices.
# El Fancy Indexing permite seleccionar un subconjunto de datos eligiendo elementos no contiguos (saltados o desordenados), a diferencia de la indexaci√≥n b√°sica que solo selecciona rangos continuos.
```

```python
## üüß 5. Producto Hadamard (multiplicaci√≥n elemento a elemento)

u = np.array([1, 2])
v = np.array([3, 2])
z = u * v     # array([3, 4])
print(z)
```

```python
## üü® 6. Producto punto (dot product) (resultado = escalar)
# Es equivalente a la suma de los productos elemento a elemento.
# En √°lgebra lineal se denota como u¬∑v o ‚ü®u, v‚ü©
# en √°lgebra lineal, el producto punto es simplemente un caso especial del producto matricial.
# se diferencia del producto Hadamard en que el resultado en el producto punto es un escalar.
u = np.array([1, 2])
v = np.array([3, 2])
# Internamente NumPy interpreta cada vector como una matriz 1√ón y n√ó1.
z = u @ v   
# (1*3 + 2*2) = 7
print(z) # 7
# z = np.dot(u, v)  
```

```python
### üü© 7. Multiplicaci√≥n de matrices (resultado = matriz ‚Üí transformaci√≥n lineal)
# Sea A una matriz de forma (m √ó n) y B una matriz de forma (n √ó p), entonces el producto matricial A @ B es una matriz de forma (m √ó p).
# A tiene forma (m √ó n)
# B tiene forma (n √ó p)
# A @ B #‚Üí matriz de forma (m √ó p)
# El primer paso que debes comprobar antes de realizar la multiplicaci√≥n matricial entre dos matrices A y B de NumPy es:
# El n√∫mero de columnas de A debe ser igual al n√∫mero de filas de B. (Requisito de Compatibilidad Dimensional)
# Ejemplo:  # (3√ó2) @ (2√ó3)
A = np.array([[1, 2],
                [3, 4],
                [5, 6]])  # Matriz A de forma (3 √ó 2)   
B = np.array([[7, 8, 9],
                [10, 11, 12]])  # Matriz B de forma (2 √ó 3)
C = A @ B  # Multiplicaci√≥n matricial A @ B
print(C)
# Resultado:
# [[ 27  30  33]
#  [ 61  68  75]
#  [ 95 106 117]]
# Explicaci√≥n del c√°lculo:
# Primera fila de A por cada columna de B:
# 1*7 + 2*10 = 27
# 1*8 + 2*11 = 30
# 1*9 + 2*12 = 33
# Segunda fila de A por cada columna de B:
# 3*7 + 4*10 = 61
# 3*8 + 4*11 = 68
# 3*9 + 4*12 = 75
# Tercera fila de A por cada columna de B:
# 5*7 + 6*10 = 95
# 5*8 + 6*11 = 106
# 5*9 + 6*12 = 117
# Por lo tanto, la matriz resultante C es:
# [[ 27  30  33]
#  [ 61  68  75]
#  [ 95 106 117]]
```
