```python
# DataFrame a partir de un diccionario 
import pandas as pd
data_dict = {
    'Producto': ['Camisa', 'Pantalón', 'Zapatos'],
    'Precio': [29.99, 49.99, 89.99],
    'Cantidad': [100, 150, 200]
}
df_dict = pd.DataFrame(data_dict)
print(df_dict)
print("------------------ Lista de productos:")
print(df_dict[['Producto']])
print("------------------ Inventario:")
print(df_dict[['Producto', 'Cantidad']])
print("------------------ uso de iloc:")
print("slicing, Todas las filas, Columnas 0 y 1:")
print(df_dict.iloc[:, [0, 1]]) 
# :  -> Todas las Filas
# [0, 1] -> Columnas 0 y 1
print("------------------ uso de iloc:")
print("nuevo dataframe, de los primeros 2 productos (mediante indice), con producto y precio:")
new_df = df_dict.iloc[0:2, [0, 1]]
print(new_df)
print("nuevo dataframe, de los primeros 2 productos (mediante tag), con producto y precio:")
new_df_loc = df_dict.loc[0:1, ['Producto', 'Precio']]
print(new_df_loc)
print("------------------ uso de iloc / loc:")
print("Selección específica:")
print(f"F2C0 -> {df_dict.iloc[2, 0]}")
print(f"F2C0 -> {df_dict.loc[2, 'Producto']}") # valido
```
