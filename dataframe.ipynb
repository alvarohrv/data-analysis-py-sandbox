```python
import pandas as pd
print("‚úÖ Pandas instalado!")
print("Versi√≥n:", pd.__version__)

# Creating a DataFrame from a dictionary
# Empresa de sofware
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack', 'Kathy', 'Liam', 'Mona', 'Nate', 'Olivia', 'Paul', 'Quinn', 'Rachel', 'Sam', 'Tina','Sonia'],
        'Age': [25, 30, 35, 28, 22, 40, 29, 31, 27, 33, 26, 24, 32, 42, 30, 28, 27, 31, 25, 26, 39],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Miami', 'Seattle', 'Boston', 'Denver', 'Austin', 'Houston', 'Philadelphia', 'Phoenix', 'San Diego', 'Dallas', 'Chicago', 'Detroit', 'Columbus', 'Chicago', 'Indianapolis', 'Miami', 'Miami'],
        'Occupation': ['Frontend', 'Designer', 'owner', 'Developer-back', 'Intern', 'Frontend', 'Developer-back', 'Designer', 'Intern', 'Developer-movil', 'Developer-movil', 'sales', 'Developer-lead', 'Intern', 'Manager', 'fullstack', 'sales', 'Intern', 'marketing', 'Manager', 'Developer-back'],
        'Salary': [70000, 80000, 120000, 95000, 40000, 72000, 88000, 60000, 45000, 98000, 91000, 65000, 115000, 42000, 130000, 105000, 68000, 43000, 75000, 125000, 90000]
        }
df = pd.DataFrame(data)
print(df)
print(f"Type::: {type(df)}")
```

### SELECCI√ìN DE DATOS EN PANDAS MEDIANTE `loc`, `iloc` Y NOMBRE DIRECTO #### 

```python
## 1. Selecci√≥n por POSICI√ìN ‚Üí `iloc`
# Se usa cuando quieres acceder por _√≠ndices num√©ricos_ (0, 1, 2‚Ä¶).
print("---- Fila 0, Columna 2")
print(df.iloc[0, 2])        # Fila 0, Columna 2 #‚Üí 'New York'
print("---- Fila 3 completa")
# Fila 3 completa, presentada de forma tubular
F3= df.iloc[3] # df.iloc[3, :] 
# type(F3)  #‚Üí pandas.core.series.Series
print(F3)
print("---- Columna 'Occupation'")
print(df.iloc[:, 3])    # Toda la columna 3 (Occupation)
print("---- Primeras 5 filas, primeras 3 columnas")
print(df.iloc[0:5, 0:3])  # Primeras 5 filas, primeras 3 columnas
print("---- Primeras 5 filas, columnas 0 y 3")
print(df.iloc[0:5, [0, 3]])
print("---- Filas 0, 5 y 15; Columnas 0, 2 y 3")
print(df.iloc[[0, 5, 15], [0, 2, 3]])  # Filas 0, 2 y 4; Columnas 1 y 3
```

```python
## 2. Selecci√≥n por POSICI√ìN ‚Üí `loc`
# Se usa cuando quieres acceder por _etiquetas de fila y/o columna.
print("---- Fila 0, Columna 'City'")
print(df.loc[0, "City"])        # Fila 0, Columna 'City' #‚Üí 'New York'
print("---- Fila 3 completa")
# Fila 3 completa, presentada de forma tubular?
print("---- Fila 3 completa")
F3 = df.loc[3]  # df.loc[3, :]
# Nota en algunos diccionarios los indices pueden ser cadenas de texto
# type(F3)  #‚Üí pandas.core.series.Series
print(F3)
print("---- Columna 'Occupation'")
print(df.loc[:, "Occupation"])    # Toda la columna 3 (Occupation)
print("---- Primeras 5 filas, primeras 3 columnas")
print(df.loc[0:4, ["Name", "Age", "City"]])
print("---- Primeras 5 filas, columnas 0 y 3")
print(df.loc[0:4, ["Name", "Occupation"]])
print("---- Filas 0, 5 y 15; Columnas 0, 2 y 3")
print(df.loc[[0, 5, 15], ["Name", "City", "Occupation"]])  # Filas 0, 2 y 4; Columnas 1 y 3
```

```python
# df[...] est√° dise√±ado para aceptar distintos tipos de argumentos, y pandas decide qu√© hacer seg√∫n el tipo de ese argumento.
# df["Name"] ‚Üí una sola etiqueta de columna ‚Üí devuelve una Series con esa columna. (selecci√≥n simple)
# df[["Name", "Occupation"]] ‚Üí lista de etiquetas de columna ‚Üí devuelve un DataFrame con esas columnas.
# df[mask_booleana] ‚Üí una m√°scara booleana ‚Üí devuelve un DataFrame filtrado seg√∫n la m√°scara.
### df[mask][['col1', 'col2']] // Indexaci√≥n directa
```

```python
# 3. Selecci√≥n simple por NOMBRE DIRECTO (solo columna) parte1
# Se usa cuando quieres acceder por _nombres de columna_ directamente.
### Para obtener **Series** (una sola columna) uso de corchetes simples `[]`
### NOTA: Para obtener **DataFrame** (varias columnas) uso de corchetes dobles

##### output: SERIES
print("---- Columna 'City' directamente")
print(df["City"])    # Toda la columna 'City'
print(type(df["City"]))  #‚Üí pandas.core.series.Series
##### output: DATAFRAME
print("---- Columna 'City' directamente como DataFrame (con cabecera)")
print(df[["City"]])    # Toda la columna 'City'
print(type(df[["City"]]))  #‚Üí pandas.core.frame.DataFrame
```

```python
# La selecci√≥n simple por NOMBRE DIRECTO permite filtrar filas basadas en condiciones espec√≠ficas aplicadas a las columnas del DataFrame.
# Esto se logra creando m√°scaras booleanas que representan las condiciones deseadas y luego aplic√°ndolas al DataFrame para obtener un subconjunto filtrado.
### Ejemplo 1
print(df[df['Age'] >= 39])  # Rows where Age >= 39
print("-----------------------------------------")

### Ejemplo 2 - M√∫ltiples condiciones:
# 1. condici√≥n 2: Filtrar por edad mayor a 28
condicion_edad = df['Age'] >= 28
# 2. condici√≥n 1: Filtrar por ciudades 'Chicago' o 'Miami'
condicion_ciudades = (df['City'] == 'Chicago') | (df['City'] == 'Miami')
# 3. condici√≥n 3: Filtrar por ocupaci√≥n 'Intern' o 'Developer 
condicion_ocupacion = (df['Occupation'] == 'Intern') | (df['Occupation'] == 'Developer')
# 4. Combina las tres condiciones con AND (&)
# (Ciudades Y Edad Y Ocupaci√≥n)
resultado = df[condicion_ciudades & condicion_edad & condicion_ocupacion][['Name', 'City', 'Occupation']]
print(resultado)
type(resultado)  #‚Üí pandas.core.frame.DataFrame
```

```python
# 3. Selecci√≥n por lista de NOMBRES DIRECTOS (solo columnas) parte2
# Se usa cuando quieres acceder por _nombres de columna_ directamente.
### Para obtener **DataFrame** (varias columnas) uso de corchetes dobles
### NOTA: Para obtener **Series** (una sola columna) uso de corchetes simples `[]`
##### output: DataFrame
#print(df[['Name']])  # Select specific columns (retorna un DataFrame y no una Serie)
print("----DataFrame de 2 columnas----")
print(df[["Name", "Occupation"]])  # Columnas 'Name' y 'Occupation'
# equivalente a: df.loc[:, ["Name", "Occupation"]]
# usa el operador de indexaci√≥n del DataFrame y solo permite seleccionar columnas por nombre cuando se pasa una lista.
#////
# NOTA: para seleccionar a la vez filas y columnas por sus ‚Äúdos ejes‚Äù en la forma df[filas, columnas] se requiere usar uno de los indexadores, iloc o loc, o bien una variante equivalente como df.where o boolean indexing, pero ya no sirve la sintaxis simple df[...] de una sola dimensi√≥n.
```

### Selecci√≥n avanzada ####

```python
# üü£ 4. Selecci√≥n avanzada
### ‚úî Indexaci√≥n directa - df[mask][['col1', 'col2']]
### De la forma:
# mascara_booleana = (condicion1) & (condicion2)
# df_filtrado = df[mascara_booleana][["columna"]]
# ----
# print(df[df['Age'] > 28])  # Rows where Age > 28
# print(df[(df['Age'] > 28) & (df['City'] == 'Chicago')])  # Multiple conditions
df[(df['Age'] > 28) & ((df['City'] == 'Chicago') | (df['City'] == 'Miami'))][['Name', 'City', 'Occupation']]
# nota: el operador & representa "AND" y el operador | representa "OR" en pandas.
# nota: la expresion [['Name', 'City', 'Occupation']] se utiliza para seleccionar m√∫ltiples columnas espec√≠ficas del DataFrame resultante despu√©s de aplicar las condiciones de filtrado.
# nota: si no se incluye esta parte, el resultado incluir√° todos lo campos(columnas) del DataFrame original para las filas que cumplen las condiciones de filtrado.
```

```python

##### CON 3 CONDICIONES #####
condicion_edad = df['Age'] > 28
condicion_ciudades = (df['City'] == 'Chicago') | (df['City'] == 'Miami')
condicion_ocupation = df['Occupation'] == 'Developer-back'
df[condicion_edad & condicion_ciudades & condicion_ocupation][['Name', 'City', 'Occupation']]
```

```python
# üü£ 4. Selecci√≥n avanzada
### ‚úî Usar `isin`
# 'isin' permite filtrar filas seg√∫n si una columna tiene valores dentro de una lista dada.
# En Excel es similar a: aplicar Filtro a la tabla por varios valores,
# o usar funciones como OR() y COUNTIF() para comprobar si un valor est√° en una lista.
#### CON 1 CONDICI√ìN #####
print(df[df['City'].isin(['Chicago', 'Miami'])]) 
# Dataframe, con las filas que tengas comoCities Chicago or Miami
```

```python
##### CON 2 CONDICIONES ##### (Es posible indicar que columnas incluir en el resultado)
# 1. Crea la condici√≥n de Ciudades usando .isin()
condicion_ciudades = df['City'].isin(['Chicago', 'Miami'])
# 2. Crea la condici√≥n de Edad
condicion_edad = df['Age'] > 28
# 3. Combina y aplica las condiciones
df[condicion_ciudades & condicion_edad][['Name', 'City', 'Occupation']]

```

```python
##### CON 3 CONDICIONES #####
# 1. Condici√≥n de Ciudades (OR)
condicion_ciudades = df['City'].isin(['Chicago', 'Miami'])
# 2. Condici√≥n de Edad (AND)
condicion_edad = df['Age'] > 28
# 3. Condici√≥n de Ocupaci√≥n (AND)
condicion_ocupacion = df['Occupation'] == 'Developer-back' # Nueva condici√≥n
# 4. Combina las tres condiciones con AND (&)
# (Ciudades Y Edad Y Ocupaci√≥n)
resultado = df[condicion_ciudades & condicion_edad & condicion_ocupacion][['Name', 'City', 'Occupation']]
print(resultado)
type(resultado)  #‚Üí pandas.core.frame.DataFrame
```

```python
# üü£ 4. Selecci√≥n avanzada
### ‚úî Selecci√≥n por query
### (filtrado con query o filtrado con expresi√≥n tipo SQL.)
# todas las columnas de los registros filtrados
# print(df.query("Age > 28 and City == 'Chicago'"))
# solo la columna name y city de los registros filtrados
# df.query("Age > 28 and City == 'Chicago'")[['Name', 'City', 'Occupation']]
df.query("Age > 28 and (City == 'Chicago' or City == 'Miami')")[['Name', 'City', 'Occupation']]
```

```python
resultado = df.query(
    "(City == 'Chicago' or City == 'Miami')"
    " and Age > 28"
    " and Occupation == 'Developer-back'"
)[['Name', 'City', 'Occupation']]
print(resultado)
```

```python
# üü£ 4. Selecci√≥n avanzada
### ‚úî loc con m√°scara booleana
##### CON 2 CONDICIONES #####
# Using .loc for label-based indexing with multiple conditions (forma Logica)
df.loc[(df['Age'] > 28) & (df['City'].isin(['Chicago', 'Miami'])), ['Name', 'City', 'Occupation']]
# nota: el operador & se usa para AND, y el operador | se usa para OR
```

```python
# Suma de salarios totales, con formato de decimales y separadores de miles
total_salary = df['Salary'].sum()
print(f"Suma total de salarios: {total_salary:,.2f}")
# (usa la coma para miles y el punto para decimales)

# promedio entre todos los empleados
average_salary = df['Salary'].mean()
# .mean() calcula el promedio aritm√©tico de los valores en la columna 'Salary'.
# los valores NaN se ignoran autom√°ticamente en el c√°lculo del promedio.
print(f"Salario promedio: {average_salary:,.2f} (21 empleados)")

```
